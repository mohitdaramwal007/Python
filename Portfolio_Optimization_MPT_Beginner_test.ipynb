📈 Portfolio Optimization with Modern Portfolio Theory (MPT)
Welcome! In this project, we’ll build an optimized stock portfolio using Modern Portfolio Theory.

This notebook is designed for beginners, with step-by-step explanations.

# 🎯 Objective

We aim to:

Download stock data
Calculate returns and risk
Simulate portfolios
Identify the best one using the Sharpe Ratio

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from scipy.optimize import minimize

%matplotlib inline
```

# 💼 Step 1: Choose Your Stocks

```python
# Added NVDA and TSLA to the portfolio to explore performance of tech and EV sector stocks.
tickers = ['MCX.NS', 'NH.NS', 'PATANJALI.NS', 'BLUEDART.NS']  # You can customize this list
```

# 📥 Step 2: Download Historical Data

```python
data = yf.download(tickers, start='2024-01-01', end='2026-06-07', auto_adjust=False)
data = data['Adj Close']
```

# 🔁 Step 3: Calculate Daily Returns

```python
# Calculate daily returns from the price data
returns = data.pct_change().dropna()
returns.head()
```

# 📊 Step 4: Calculate Mean Returns and Covariance

```python
# Calculate mean returns and covariance matrix
mean_returns = returns.mean()
cov_matrix = returns.cov()

# Display them
print("Mean Daily Returns:")
print(mean_returns)

print("\nCovariance Matrix:")
print(cov_matrix)
```

# ⚙️ Step 5: Define Portfolio Performance Functions

```python
import numpy as np

def portfolio_performance(weights, mean_returns, cov_matrix):
    returns = np.dot(weights, mean_returns)
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe_ratio = returns / std
    return returns, std, sharpe_ratio
```

# 🚧 Step 6: Optimization Constraints

```python
from scipy.optimize import minimize

num_assets = len(tickers)
args = (mean_returns, cov_matrix)

# Constraint: sum of weights = 1 (i.e., 100% of investment)
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Bounds: weights between 0 and 1 (i.e., no short selling)
bounds = tuple((0, 1) for _ in range(num_assets))
```

# 🚀 Step 7: Optimize the Portfolio

```python
# Initial guess: equal allocation
initial_weights = num_assets * [1. / num_assets]

# Function to minimize (negative Sharpe ratio)
def neg_sharpe_ratio(weights, mean_returns, cov_matrix):
    return -portfolio_performance(weights, mean_returns, cov_matrix)[2]

# Optimization using 'SLSQP' method
optimized = minimize(neg_sharpe_ratio, initial_weights,
                     args=args, method='SLSQP',
                     bounds=bounds, constraints=constraints)

# Get optimal weights
optimal_weights = optimized.x
```

# ✅ Step 8: Show Optimized Portfolio

```python

ret, vol, sharpe = portfolio_performance(optimal_weights, mean_returns, cov_matrix)

print(f"\nExpected Annual Return: {ret * 252:.2f}")
print(f"Annual Volatility: {vol * np.sqrt(252):.2f}")
print(f"Sharpe Ratio: {sharpe:.2f}")
```
